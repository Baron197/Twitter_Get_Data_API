```python
import pandas as pd
import json
```

```python
tweetList = []
```

```python
with open('google.json') as json_file:  
    data = json.load(json_file)
   
    for item in data :
        hashtags = ''
        for hashtag in item['entities']['hashtags']:
            hashtags += hashtag['text'] + ' ';
        theTweet = item['text']
        if('extended_tweet' in item.keys()) :
            theTweet = item['extended_tweet']['full_text']
        tweetList.append([item['id'], 
                          item['created_at'], 
                          theTweet,
                          item['user']['id'],
                          item['user']['name'],
                          item['user']['screen_name'],
                          hashtags,
                          item['source']])
        
dfStream = pd.DataFrame(tweetList, columns=['id','createdDate',
                                            'text','userId',
                                            'userName','userAccount',
                                            'hashtags','source'])
dfStream
```

```python
dfStream.loc[7,'text']
```

```python
tweetList = []
with open('scoopy.json') as json_file:  
    data = json.load(json_file)
    listTweets = data['statuses']
    for item in listTweets :
        hashtags = ''
        for hashtag in item['entities']['hashtags']:
            hashtags += hashtag['text'] + ' ';
        theTweet = item['text']
        if('extended_tweet' in item.keys()) :
            theTweet = item['extended_tweet']['full_text']
        tweetList.append([item['id'], 
                          item['created_at'], 
                          theTweet,
                          item['user']['id'],
                          item['user']['name'],
                          item['user']['screen_name'],
                          hashtags])
dfSearchScoopy = pd.DataFrame(tweetList, columns=['id','createdDate','text',
                                                 'userId','userName','userAccount',
                                                 'hashtags'])
dfSearchScoopy
```
